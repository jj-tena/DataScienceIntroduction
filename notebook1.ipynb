# Visualización de datos
El objetivo de este cuarderno es mostrar una serie de gráficas para un dataset generadas mediante las librerías Pandas, Matplotlib y Seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

data = {
    'A': np.random.randint(1, 11, size=10),
    'B': np.random.randint(1, 11, size=10),
    'C': np.random.randint(1, 11, size=10),
    'D': np.random.choice(['Rojo', 'Verde', 'Azul'], size=10)
}

df = pd.DataFrame(data)

print(df)
```

# Pandas

Gráficas generadas mediante la librería Pandas




```python
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

df.plot.scatter(x='A', y='B', ax=axs[0, 0])
axs[0, 0].set_title('Scatter Plot (A vs B)')

df['C'].plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Bar Plot (C)')

df['A'].plot(kind='hist', bins=5, ax=axs[0, 2])
axs[0, 2].set_title('Histogram (A)')

df['D'].value_counts().plot.pie(autopct='%1.1f%%', ax=axs[1, 0])
axs[1, 0].set_title('Pie Chart (D)')

df['B'].plot(kind='box', ax=axs[1, 1])
axs[1, 1].set_title('Box Plot (B)')

df[['A', 'C']].plot(ax=axs[1, 2])
axs[1, 2].set_title('Line Plot (A and C)')

df['B'].plot(kind='hist', bins=10, ax=axs[2, 0])
axs[2, 0].set_title('Histogram (B)')

df[['A', 'B', 'C']].plot.area(ax=axs[2, 1])
axs[2, 1].set_title('Stacked Area Plot (A, B, C)')

df.plot.line(ax=axs[2, 2])
axs[2, 2].set_title('Line Plot (All Columns)')

plt.tight_layout()
plt.show()
```

# Matplotlib
Gráficas generadas mediante la librería Matplotlib


```python
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

axs[0, 0].scatter(df['A'], df['B'])
axs[0, 0].set_title('Scatter Plot (A vs B)')

axs[0, 1].bar(df.index, df['C'])
axs[0, 1].set_title('Bar Plot (C)')

axs[0, 2].hist(df['A'], bins=5)
axs[0, 2].set_title('Histogram (A)')

axs[1, 0].pie(df['D'].value_counts(), labels=df['D'].unique(), autopct='%1.1f%%')
axs[1, 0].set_title('Pie Chart (D)')

axs[1, 1].boxplot(df['B'])
axs[1, 1].set_title('Box Plot (B)')

axs[1, 2].plot(df[['A', 'C']])
axs[1, 2].legend(['A', 'C'])
axs[1, 2].set_title('Line Plot (A and C)')

axs[2, 0].hist(df['B'], bins=10)
axs[2, 0].set_title('Histogram (B)')

axs[2, 1].fill_between(df.index, df['A'], df['B'], alpha=0.6)
axs[2, 1].set_title('Stacked Area Plot (A and B')

axs[2, 2].plot(df[['A', 'B', 'C']])
axs[2, 2].legend(['A', 'B', 'C'])
axs[2, 2].set_title('Line Plot (All Columns)')

plt.tight_layout()
plt.show()
```

# Seaborn
Gráficas generadas mediante la librería Seaborn


```python
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

sns.scatterplot(x='A', y='B', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Scatter Plot (A vs B)')

sns.barplot(x=df.index, y='C', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Bar Plot (C)')

sns.histplot(df['A'], bins=5, kde=True, ax=axs[0, 2])
axs[0, 2].set_title('Histogram (A)')

df['D'].value_counts().plot.pie(autopct='%1.1f%%', ax=axs[1, 0])
axs[1, 0].set_title('Pie Chart (D)')

sns.boxplot(x='B', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Box Plot (B)')

sns.lineplot(data=df[['A', 'C']], ax=axs[1, 2])
axs[1, 2].set_title('Line Plot (A and C)')

sns.violinplot(x='B', data=df, ax=axs[2, 0])
axs[2, 0].set_title('Violin Plot (B)')

sns.lineplot(data=df[['A', 'B', 'C']], ax=axs[2, 1])
axs[2, 1].set_title('Line Plot (A, B, C)')

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', ax=axs[2, 2])
axs[2, 2].set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```
